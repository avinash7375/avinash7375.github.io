```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.rottentomatoes.com/top/bestofrt/"
# Changing user agent (some sites block default python user agent)
header = {"User-Agent": "My User Agent"}

f = requests.get(url, headers=header)

soup = BeautifulSoup(f.content, "html.parser")

# selecting the div that contains all the movies
movies_div = soup.find("div", {"class": "discovery-tiles__wrap"})

movies = []

for a_tag in movies_div.find_all("a"):
    movies.append(a_tag["href"].strip())

movies_link = []
movies_synopsis = []
movies_names = []

for movie in movies:
    url = "https://www.rottentomatoes.com" + movie
    movies_link.append(url.strip())

    movie_html = requests.get(url, headers=header)
    movie_soup = BeautifulSoup(movie_html.content, "html.parser")
    movie_content = movie_soup.find("p", {"data-qa": "movie-info-synopsis", "slot": "content"})
    movie_name = movie_soup.find("h1", {"data-qa": "score-panel-title"})



    if movie_name:
        movies_names.append(movie_name.string.strip())
    else:
        movies_names.append("-")

    if movie_content:
        movies_synopsis.append(movie_content.string.strip())
    else:
        movies_synopsis.append("-")

df = pd.DataFrame(
    {
        "Movie Name": movies_names,
        "Movie Link": movies_link,
        "Movie Synopsis": movies_synopsis,
    }
)
print(df.head())
df.to_csv("./movies.csv")
```
