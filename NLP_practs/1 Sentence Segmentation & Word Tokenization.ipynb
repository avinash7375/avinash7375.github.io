```python
# AIM: Write a program to implement Sentence Segmentation & Word Tokenization
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
#words in file are "This is start of new Era"
with open("hello.txt")as sente:
    lines = sente.readlines()
```

```python
for content in lines:
    line = nltk.sent_tokenize(content)
    print(content)
    print("Tokens:", word_tokenize(content))
```
